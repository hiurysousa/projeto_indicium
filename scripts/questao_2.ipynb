```python
import pandas as pd
from funcoes.tratamento import tratar_categoria
```

```python
caminho_produto = 'C:/Users/hiury/OneDrive/Área de Trabalho/IFCE/projeto_indicium/datasets/raw/produtos_raw.csv'
df_produto = pd.read_csv(caminho_produto)
```

```python
df_produto['actual_category'] = df_produto['actual_category'].apply(
    tratar_categoria, args=('ele', 'eletrônicos')
)

df_produto['actual_category'] = df_produto['actual_category'].apply(
    tratar_categoria, args=('prop','propulsão')
)

df_produto['actual_category'] = df_produto['actual_category'].apply(
    tratar_categoria, args=('cor', 'ancoragem')
)
```

```python
if df_produto['price'].str.contains('R\$').any():
    df_produto['price'] = df_produto['price'].str.replace('R$', '', regex=False).str.strip()

df_produto['price'] = df_produto['price'].astype(float)
```

```python
print(f'Total de duplicatas: {df_produto.duplicated().sum()}')
df_produto = df_produto.drop_duplicates()
print('\nRemovendo duplicatas!\n')
print(f'Total de duplicatas: {df_produto.duplicated().sum()}')
```

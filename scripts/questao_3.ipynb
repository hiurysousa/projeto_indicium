```python
import pandas as pd
```

```python
caminho_importacao = 'C:/Users/hiury/OneDrive/Área de Trabalho/IFCE/projeto_indicium/datasets/raw/custos_importacao.json'
df_importacao = pd.read_json(caminho_importacao)
```

```python
df_importacao.head()
```

```python
df_importacao_explodido = df_importacao.explode('historic_data')
pd.set_option('display.max_colwidth', None)
df_importacao_explodido
```

```python
df_historico = pd.json_normalize(df_importacao_explodido['historic_data'])
df_importacao_explodido = df_importacao_explodido.reset_index(drop=True)
df_importacao_final = pd.concat([
    df_importacao_explodido[['product_id', 'product_name', 'category']],
    df_historico
], axis=1)

df_importacao_final
```

```python
try:
    df_importacao_final.to_csv('C:/Users/hiury/OneDrive/Área de Trabalho/IFCE/projeto_indicium/datasets/processed/df_importacao_final.csv', index=False)
    print(f'Arquivo criado com sucesso!')
except Exception as e:
    print(f'Erro! {e}')

print(f'Antes da normalização: {df_importacao.shape} | Depois: {df_importacao_final.shape}')
```
